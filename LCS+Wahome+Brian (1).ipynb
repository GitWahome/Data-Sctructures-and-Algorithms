<h4>1. &nbsp; Write python code to give the length of the longest common subsequence for two strings.</h4>

</div>

```python
from itertools import combinations
def matGen(string1, string2):
    """
    -In my code, I prefered not to generate a matrix of arrows rather, reverse engineer
     the conditionals to determine the order of reading our LCS.
    -This idea was based on
     https://stackoverflow.com/questions/32288223/dynamic-programming-longest-common-subsequence#32290433
    -The whole however code is entirely written by myself.
    """
    matrix = [[0 for j in range(len(string2) + 1)] for i in range(len(string1) + 1)]
    for i, x in enumerate(string1):
        for j, y in enumerate(string2):
            if x == y:
                matrix[i+1][j+1] = matrix[i][j] + 1
            else:
                matrix[i+1][j+1] = max(matrix[i+1][j], matrix[i][j+1])
    return matrix
def  backTrack(matrix, string1, string2):
    result = []
    ls1, ls2 = len(string1), len(string2)
    while ls1 != 0 and ls2 != 0:
        if matrix[ls1][ls2] == matrix[ls1-1][ls2]:
            ls1 -= 1
        elif matrix[ls1][ls2] == matrix[ls1][ls2-1]:
            ls2 -= 1
        else:
            if string1[ls1 -1] == string2[ls2-1]:
                result.append(string1[ls1 - 1])
                ls1 -= 1
                ls2 -= 1
    return "".join([ls1 for ls1 in reversed(result)])
def lcs(string1,string2):
    #Improvement
    if string1==string2:
        return string2
    else:
        return (backTrack(matGen(string1,string2),string1,string2))
```

```python
DNA=[(0,'TTCTACGGGGGGAGACCTTTACGAATCACACCGGTCTTCTTTGTTCTAGCCGCTCTTTTTCATCAGTTGCAGCTAGTGCATAATTGCTCACAAACGTATC'),
     (1,'TCTACGGGGGGCGTCATTACGGAATCCACACAGGTCGTTATGTTCATCTGTCTCTTTTCACAGTTGCGGCTTGTGCATAATGCTCACGAACGTATC'),
     (2,'TCTACGGGGGGCGTCTATTACGTCGCCAACAGGTCGTATGTTCATTGTCATCATTTTCATAGTTGCGGCCTGTGCGTGCTTACGAACGTATTCC'),
     (3,'TCCTAACGGGTAGTGTCATACGGAATCGACACGAGGTCGTATCTTCAATTGTCTCTTCACAGTTGCGGCTGTCCATAAACGCGTCCCGAACGTTATG'),
     (4,'TATCAGTAGGGCATACTTGTACGACATTCCCCGGATAGCCACTTTTTTCCTACCCGTCTCTTTTTCTGACCCGTTCCAGCTGATAAGTCTGATGACTC'),
     (5,'TAATCTATAGCATACTTTACGAACTACCCCGGTCCACGTTTTTCCTCGTCTTCTTTCGCTCGATAGCCATGGTAACTTCTACAAAGTTC'),
     (6,'TATCATAGGGCATACTTTTACGAACTCCCCGGTGCACTTTTTTCCTACCGCTCTTTTTCGACTCGTTGCAGCCATGATAACTGCTACAAACTTC')]
def dnAnalytics():
    listIndsDNA=[x for x in range(0,len(DNA))]
    stringCombos=[x for x in combinations(listIndsDNA,2)]
    #print(stringCombos)
    generations=0
    analysisSequences={}
    analysisCombos={}
    insertions,deletions=0,0
    for combos in stringCombos:
        #print(combos[0])
        insertions,deletions=0,0
        inheritedDNA=lcs(DNA[combos[0]][1],DNA[combos[1]][1])
        insDel=(len(DNA[combos[0]][1])-len(inheritedDNA))-(len(DNA[combos[1]][1])-len(inheritedDNA))
        if insDel>=0:
            insertions=insDel
        elif insDel<0:
            deletions=abs(insDel)
        mutations=(len(DNA[combos[1]][1])-len(inheritedDNA))-abs(insDel)
        #deletions=len(DNA[generations][1])-len(inheritedDNA)
        probsMutation=float(mutations)/len(DNA[combos[0]][1])
        probsInsertion=float(insertions)/len(DNA[combos[0]][1])
        probsDeletion=float(deletions)/len(DNA[combos[0]][1])
        analysisSequences["{} vs {}".format(DNA[combos[0]][0],DNA[combos[1]][0])]={
            "Mutations":mutations,
            "Length LCS":len(inheritedDNA),
            "LCS":inheritedDNA,
            "Insertions":insertions,
            "Deletions":deletions,
            "Probability of Mutation":probsMutation,
            "Probability of Insertion":probsInsertion,
            "Probability of Deletion":probsDeletion
        }
        
    return {"Data":analysisSequences,"Combinations":stringCombos}
def tablePrintLCSlen():
    data=dnAnalytics()
    print("________________________________________________________________________________________")
    for vals in range(len(data['Combinations'])):
        combination=" vs ".join(([str(x) for x in data['Combinations'][vals]]))
        print("String pair: {}| Length LCS:{} \n"
              .format(combination,
                      data['Data'][combination]["Length LCS"]))
        print("________________________________________________________________________________________")
def tablePrintLCSProbs():
    data=dnAnalytics()
    for vals in range(len(data['Combinations'])):
        combination=" vs ".join(([str(x) for x in data['Combinations'][vals]]))
        print("**************************************************************************************")
        print("Probability Mutation: String pair: {}| Length LCS:{} \n"
              .format(combination,
                      data['Data'][combination]["Probability of Mutation"]))
        print("Probability Insertion: String pair: {}| Length LCS:{} \n"
              .format(combination,
                      data['Data'][combination]["Probability of Insertion"]))
        print("Probability Deletion: String pair: {} | Length LCS:{} \n"
              .format(combination,
                      data['Data'][combination]["Probability of Deletion"]))
        print("________________________________________________________________________________________")
        
```
